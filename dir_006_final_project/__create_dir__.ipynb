```python
import os
import re
import numpy as np
```

```python
# read file
open_file = open("./project_ideas.md", "r")
read_file = open_file.read()
read_file
```

```python
# split file by new line
split_file = read_file.split("\n")
split_file


# array that exist ". " in the line
arrays = []
for i in split_file:
    if "- [ ]" in i:
        arrays.append(i)

arrays
```

```python
new_arrays = arrays.copy()

new_arrays = [array.lower() for array in new_arrays]

# new_arrays = [re.search(r"\. (.*)", array).group(1) if re.search(r"\. (.*)", array) else array for array in new_arrays]


new_arrays = [array.replace("- [ ] ", "") for array in new_arrays]

new_arrays = [array.replace(".", "_") for array in new_arrays]

new_arrays = [array.replace(" ", "_") for array in new_arrays]

new_arrays = [array.replace("-", "_") for array in new_arrays]

# new_arrays = [f"idea_{i+1:03d}_" + array for i, array in enumerate(new_arrays)]
new_arrays = [f"proj_{i+1:03d}_" + array for i, array in enumerate(new_arrays)]

new_arrays
```

```python
# create folder
for i in new_arrays:
    os.makedirs(i, exist_ok=True)
```
